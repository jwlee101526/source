### 튜플
- ()
- 튜플은 리스트와 유사하나, 요소값을 변경할 수 없다.

```python
t1 = ()
t2 = (1) # 해당 값은 튜플이 아니다.
t3 = (1,) # 한 개 단위는 뒤에 콤마를 붙여 사용
t4 = (1,2,3)
t5 = ('a', 'b', ('c','d'))

print(t1)
print(t2)
print(t3)
print(t4)

# t4[0] = 0  // 불변 값이므로 수정 불가
```

### 딕셔너리 자료형
- {key1:value1, key2, value2}
- key 값은 중복 불가

```python
dict1 = {"name" : "kim", "age" : 22}
dict2 = {1:"kim", 2:22, 3:"서울"}
dict3 = {"arr" : [1,2,3,4,5]}

print(dict1["name"])
print(dict1.get("name"))
print(dict2.get(3))

dict1['birth'] = '1999-09-12' # 딕셔너리 추가
print(dict1)

print(dict1.keys())
print(dict1.values())
print(dict1.items())
```

```python
try:
    user_input = input("숫자를 입력하세요: ")
    num = int(user_input)
    
    validatation = num % 2

    match validatation:
        case 0:
            print("짝수입니다.")
        case 1:
            print("홀수입니다.")
        case _:
            print("처리할 수 없는 결과입니다.")

except ValueError:
    print(f"오류: '{user_input}'은(는) 유효한 정수가 아닙니다.")
```
