### Tool Calling
- 정의된 함수를 호출

```python
from ollama import chat

def calculator(a, b):
    return a + b

tools = [
    {
        "type": "function",
        "function": {
            "name": "calculator",
            "description": "Add two numbers.",
            "parameters": {
                "type": "object",
                "properties": {
                    "a": {"type": "integer", "description": "The first number to add."},
                    "b": {"type": "integer", "description": "The second number to add."},
                },
                "required": ["a", "b"],
            },
        },
    }
]

response = chat(
    model="qwen2.5:7b",
    messages=[
        {"role": "system", "content": "너는 계산기 도구를 사용하여 계산을 수행하는 유능한 비서야. 반드시 제공된 calculator 도구를 사용해서 계산해."},
        {"role": "user", "content": "25와 36을 더해라"}
    ],
    tools=tools,
    options={"num_ctx": 16384, "temperature": 0.1}
)

print(response)
if response.message.tool_calls:
    print("도구 호출 성공:", response.message.tool_calls)
else:
    print("도구 호출 실패. 모델 응답:", response.message.content)
```

```python
tool_call = response.message.tool_calls[0]
args = tool_call.function.arguments

result = calculator(args["a"], args["b"])

messages = [
    {"role" : "user", "content" : "25와 36을 더해라"},
    response.message,
    {"role" : "tool", "content": str(result)},
]

final_response = chat(model='qwen2.5:7b', messages=messages)
print(final_response.message.content)
```

```python
import ollama, json


# Tool 함수들
def get_weather(city):
    return {"temp": 12, "condition": "비"}


def need_umbrella(condition):
    return {"result": condition in ["비", "눈"]}


def calculate(expression):
    return {"result": eval(expression)}  # 실제론 안전한 파서 사용


# Tool 스키마 목록
tools = [
    {
        "type": "function",
        "function": {
            "name": "get_weather",
            "description": "도시 날씨 조회",
            "parameters": {
                "type": "object",
                "properties": {"city": {"type": "string"}},
                "required": ["city"],
            },
        },
    },
    {
        "type": "function",
        "function": {
            "name": "need_umbrella",
            "description": "날씨 상태로 우산 필요 여부 판단",
            "parameters": {
                "type": "object",
                "properties": {"condition": {"type": "string"}},
                "required": ["condition"],
            },
        },
    },
]

tool_map = {
    "get_weather": get_weather,
    "need_umbrella": need_umbrella,
}

messages = [{"role": "user", "content": "서울 날씨 보고 우산이 필요한지 알려줘"}]

# ── 자동 루프: Tool 호출이 없을 때까지 반복 ──
while True:
    response = ollama.chat(model="qwen2.5:7b", messages=messages, tools=tools)

    if not response.message.tool_calls:
        # Tool 호출 없음 = 최종 답변
        print(response.message.content)
        break

    messages.append(response.message)

    for tc in response.message.tool_calls:
        name = tc.function.name
        args = tc.function.arguments
        result = tool_map[name](**args)
        print(f"[Tool] {name}({args}) → {result}")

        messages.append(
            {
                "role": "tool",
                "content": json.dumps(result, ensure_ascii=False),
            }
        )
```

임베딩 값을 이용해

RAG : 주어진 문서 안에서 질의 응답 가능
분류, 군집 : 비슷한 의미의 문서끼리 그룹핑
Vector DB : 임베딩 값을 저장할 수 있는 데이터베이스

```python
from numpy import dot
from numpy.linalg import norm

def cosine_similarity(a, b):
    return dot(a, b) / (norm(a) * norm(b))
```

```python
import ollama

# 두 개의 문장이 유사한가?

s1 = '파이썬은 프로그래밍 언어이다'
s2 = 'Python은 개발에 사용되는 언어이다'

e1 = ollama.embed(model='qwen3-embedding:4b', input=s1).embeddings[0]
e2 = ollama.embed(model='qwen3-embedding:4b', input=s2).embeddings[0]

print(cosine_similarity(e1,e2))
```

```python
documents = [
    "파이썬은 프로그래밍 언어이다.",
    "축구는 세계적으로 인기있는 스포츠이다.",
    "인공지능은 머신러닝을 활용한다.",
    "서울은 대한민국의 수도이다."
]

doc_vectors = []

for doc in documents:
    vector = ollama.embed(model = 'qwen3-embedding:4b' , input=doc).embeddings[0]
    doc_vectors.append(vector)

query = "AI 기술이란?"
query_vector = ollama.embed(model = 'qwen3-embedding:4b', input=query).embeddings[0]

scores = []
for doc, vec in zip(documents, doc_vectors):
    score = cosine_similarity(query_vector, vec)
    scores.append((doc, score))

scores.sort(key=lambda x:x[1], reverse=True)

for doc, score in scores:
    print(doc, score)
```

```python
from ollama import chat

context = scores[0][0]

prompt = f"""
문서: {context}

질문: {query}
"""

response = chat(
    model = "gemma4:e2b",
    messages=[
        {"role":"user", "content": prompt}
    ]
)

print(response)
```
