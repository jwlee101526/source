```python
from dotenv import load_dotenv
import os
from ibm_watsonx_ai import APIClient
from ibm_watsonx_ai import Credentials
from ibm_watsonx_ai.foundation_models import ModelInference
```

```python
# .env 내용 가져오기
load_dotenv()

apikey = os.getenv("WATSONX_API_KEY")
project_id = os.getenv("WATSONX_PROJECT_ID")
watsonx_ai_url = os.getenv("WATSONX_URL")
```

```python
# 현재 watsonx 에서 제공하는 모델 알아보기

credentials = Credentials(
   url = f"{watsonx_ai_url}",
   api_key = f"{apikey}",
)
client = APIClient(credentials)

client.foundation_models.TextModels.show()
```

```python
credentials = Credentials(
   url = f"{watsonx_ai_url}",
   api_key = f"{apikey}",
)
client = APIClient(credentials)

model = ModelInference(
      model_id="ibm/granite-4-h-small",
      api_client=client,
      project_id=f"{project_id}",
            params = {
                  "max_tokens": 100
            }
)
# 사용자 프롬프트
user_prompt = 'How far is Paris from Bangalore?'
messages = [
    # 시스템 프롬프트 
#     {"role":"system", "content":"You are a helpful assistant."},
    {"role":"user", "content":user_prompt},
]

generated_response = model.chat(messages=messages)

print(generated_response)
print(generated_response['choices'][0]['message']['content'])
```

```python
credentials = Credentials(
   url = f"{watsonx_ai_url}",
   api_key = f"{apikey}",
)
client = APIClient(credentials)

model = ModelInference(
      model_id="ibm/granite-4-h-small",
      api_client=client,
      project_id=f"{project_id}",
            params = {
                  "max_tokens": 2000
            }
)
# 사용자 프롬프트
user_prompt = """페페로니 피자 만드는 방법 설명해줘.
"""

system_prompt = """
You are a cooking assistant.
- Answer only questions related to cooking, recipes, or ingredients.
- Provide clear and practical instructions.
- Use simple measurements and steps.
- Prefer bullet points over long paragraphs.
"""

messages = [
    # 시스템 프롬프트 
    {"role":"system", "content":system_prompt},
    {"role":"user", "content":user_prompt},
]

# generated_response = model.chat(messages=messages)

# print(generated_response)
# print(generated_response['choices'][0]['message']['content'])


generated_response_stream = model.chat_stream(messages=messages)
for chunk in generated_response_stream:
    if chunk['choices']:
            print(chunk['choices'][0]['delta'].get('content',''),end="", flush=True)
```

```python
credentials = Credentials(
   url = f"{watsonx_ai_url}",
   api_key = f"{apikey}",
)
client = APIClient(credentials)

model = ModelInference(
      model_id="ibm/granite-4-h-small",
      api_client=client,
      project_id=f"{project_id}",
            params = {
                  "max_tokens": 2000,
                  "temperature": 1.5
            }
)
# 사용자 프롬프트
user_prompt = """무서운 이야기 만들어 줄래?
"""

system_prompt = """
너는 유명한 공포 소설 작가야.
"""

messages = [
    # 시스템 프롬프트 
    {"role":"system", "content":system_prompt},
    {"role":"user", "content":user_prompt},
]

generated_response = model.chat(messages=messages)

print(generated_response)
print(generated_response['choices'][0]['message']['content'])
```

```python
print(generated_response['usage'])

# completion_tokens : 모델이 생성한 응답에 사용된 토큰 수
# prompt_tokens : 모델이 입력으로 받은 내용에 사용된 토큰 수

total = 0.0000636 / 1000 * generated_response['usage']['prompt_tokens'] + 0.000265 / 1000 * generated_response['usage']['completion_tokens']

print(total * 1490)
```

```python
import os, base64

filename = "./image/2.jpg"

if not os.path.exists(filename):
    print(f"오류: {filename} 파일이 존재하지 않습니다. 경로를 확인해주세요.")
else:
    with open(filename, "rb") as image_file:
        image_b64_encoded_string = base64.b64encode(image_file.read()).decode("utf-8")

    model = ModelInference(model_id="meta-llama/llama-3-2-11b-vision-instruct",
                           api_client=client,
                           project_id=f"{project_id}",
                           params={
                               "max_tokens": 2000
                           }
    )

    system_prompt = """
    당신은 이미지 분석 전문가 AI입니다. 사용자의 요청에 따라 
    - 이미지 설명
    - 분위기 분석
    - 감정 분석
    - 객체 설명 

    등을 수행하세요
    """

    user_prompt = "이미지를 분석해줘"

    messages = [
        {"role": "system", "content": system_prompt},
        {
            "role": "user",
            "content": [
                {
                    "type": "image_url", 
                    "image_url": {
                        "url": f"data:image/jpeg;base64,{image_b64_encoded_string}"
                        }
                    },
                {
                    "type": "text",
                    "text": user_prompt,
                },
            ],
        },
    ]

    generated_response = model.chat(messages=messages)
    print(generated_response['choices'][0]['message']['content'])
```
